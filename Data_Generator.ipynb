```python
import pandas as pd
import numpy as np

# 1. Create the Transaction Log (Sales & COGS)
np.random.seed(42)
dates = pd.date_range(start='2024-01-01', end='2025-12-31', freq='D')
products = ['Aura-Scent Diffuser', 'Lumina Smart Lamp', 'Zenith Air Enhancer']

data = []
for date in dates:
    for product in products:
        units = np.random.randint(10, 50)
        # Year 2 (2025) has slightly higher prices but higher costs (Inflation)
        base_price = 150 if date.year == 2024 else 160 
        base_cost = 70 if date.year == 2024 else 85
        
        data.append([date, product, units, base_price, base_cost])

sales_df = pd.DataFrame(data, columns=['Date', 'Product', 'Units_Sold', 'Price', 'Unit_Cost'])
sales_df.to_csv('sales_ledger.csv', index=False)

# 2. Create the Overhead Ledger (Fixed Costs)
overheads = {
    'Category': ['Salaries', 'Rent', 'Marketing', 'Utilities', 'Software'],
    'Monthly_Cost_2024': [20000, 5000, 8000, 2000, 1500],
    'Monthly_Cost_2025': [22000, 5500, 12000, 2500, 1800] # CBDO increased marketing in 2025!
}
overheads_df = pd.DataFrame(overheads)
overheads_df.to_csv('overheads.csv', index=False)

print("✅ Data Foundation Files Created: sales_ledger.csv & overheads.csv")
```
